```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_frequencies(numax, dnu, dl, c=0.0, nn=3):
    dn = np.arange(-nn, nn+1)
    dl = np.atleast_1d(dl)
    nl = len(dl)
    l = np.arange(nl)
    
    nu = numax + ((0.5*l + dl)[:, None] + (dn + c*dn**2)[None, :]) * dnu
    return nu

def get_amplitudes(nu, numax, amps, width):
    return np.atleast_1d(amps)[:, None] * np.exp(-0.5*((nu - numax)/width)**2)
```

```python
dnu = 18
numax = 220.1
nu = get_frequencies(numax, dnu, [0.0, 0.01], c=0.01)

plt.plot(nu.T % dnu, nu.T, ".-")
plt.xlim(0, dnu)
```

```python
amps = get_amplitudes(nu, numax, [100.0, 125.0], 25)
plt.plot(nu.T, amps.T, ".-")
```
